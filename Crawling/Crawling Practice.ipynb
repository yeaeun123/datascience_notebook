```python
from urllib.request import urlopen
from bs4 import BeautifulSoup

req = urlopen('https://land.naver.com/')
print("status: ", req.getcode())


if req.getcode() == 200:  # 성공
    # html 받아오기
    html = req.read()
#     print(html) # 인코딩이 깨진다.
    html = html.decode("utf-8")
    print(html)
else: # 실패
    print("HTTP-ERROR")
```

```python
# html 소스 -> soup 객체로 변환
soup = BeautifulSoup(html, "html.parser")  # soup 생성

# 문서 정보 확인
print("문서 제목: ", soup.title.text)
```
